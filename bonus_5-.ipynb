```python
balance=1000
min_balance=500

while True:
  print("select an option:")
  print(end="\n")
  print("A. Withdraw")
  print("B. Deposit")
  print("C. Check Balance")
  print("D. Exit")
  a=input("Enter an Option:")
  if a=="A":
    withdraw_amount=int(input("Enter Withdraw amount"))
    if balance>min_balance:
      if balance-withdraw_amount>=min_balance:
        print("Please take your Cash")
        balance-=withdraw_amount
      elif balance-withdraw_amount<min_balance:
        print("You are exceeding minmum balance limit")
        print("Please enter an amount maintaining minimum balance")
    elif balance <= min_balance:
      print("Insufficiunt Balance")
  elif a=="B":
    deposit_amount=int(input("Enter an amount to deposit"))
    balance+=deposit_amount
    print("Thankyou")
  elif a=="C":
    print("Available Balance is {}".format(balance))
    print("Thankyou")
  else:
    break
```

```python
password=input("Enter a strong Password ")
uppercase="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
lowercase="abcdefghijklmnopqrstuvwxyz"
num="0123456789"
special_characters="~!@#$%^&*()_-+></\'"
c=0
if len(password)>=8:
  for i in password:
    if i in uppercase:
      c+=1
      break
  for i in password:
    if i in lowercase:
      c+=1
      break
  for i in password:
    if i in special_characters:
      c+=1
      break
  for i in password:
    if i in num:
      c+=1
      break
else:
  print("This is not a Strong password")
  print("Please enter a strong password")
if c>=4:
  print("This is a strong password")
else:
  print("This is not a Strong password")
  print("Please enter a strong password")

```
