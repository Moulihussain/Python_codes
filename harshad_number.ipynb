```python
 #Harshad Number
n=int(input("Enter any Number: "))
sum_num=0
for i in str(n):
  sum_num+=int(i)
if (n%sum_num == 0):
  print("It's a Harshad Number")
else:
  print("It's not a Harshad Number")
```

```python
#Harshad Number
n=int(input())
t=n
sum_num=0
while n>0:
  sum_num+=(n%10)
  n=n//10
if t%sum_num==0:
  print("Harshad Number")
else:
  print("Not Harshad Number")
```

```python
#Auto morphic
n=int(input("Enter any Number: "))
square_num=n**2
m=len(str(n))
l=len(str(square_num))
#print((str(n))==(str(square_num)[l-m:]))
if (str(n))==(str(square_num)[l-m:]):
  print("Auto Morphic")
else:
  print("Not Auto Morphic")
```
